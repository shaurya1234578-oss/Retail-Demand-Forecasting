# Feature Engineering

```python
import pandas as pd
df = pd.read_csv('../data/sample_sales.csv', parse_dates=['date'])
df = df.sort_values(['store_id','product_id','date'])
df['day_of_week'] = df['date'].dt.dayofweek
df['month'] = df['date'].dt.month
df['lag_1_units'] = df.groupby(['store_id','product_id'])['units_sold'].shift(1)
df['lag_7_units'] = df.groupby(['store_id','product_id'])['units_sold'].shift(7)
df['rolling_7_units'] = df.groupby(['store_id','product_id'])['units_sold'].transform(lambda s: s.shift(1).rolling(7, min_periods=1).mean())
df[['lag_1_units','lag_7_units','rolling_7_units']] = df[['lag_1_units','lag_7_units','rolling_7_units']].fillna(df['units_sold'].median())
print(df[['date','store_id','product_id','units_sold','lag_1_units','rolling_7_units','snap_flag']].head(8).to_string(index=False))
```

```python
encoded = pd.get_dummies(df, columns=['store_id','product_id','category','city'])
print(encoded.shape)
print([col for col in encoded.columns if 'snap_flag' in col or 'lag_' in col])
```
