# Model Training - XGBoost vs Prophet

```python
import pandas as pd
from pathlib import Path
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

df = pd.read_csv('../data/sample_sales.csv', parse_dates=['date'])
print(df.shape)
```

```python
# XGBoost training cell. Run after installing requirements.txt.
from xgboost import XGBRegressor

feature_cols = ['sell_price', 'promotion_flag', 'snap_flag']
model = XGBRegressor(n_estimators=120, learning_rate=0.08, max_depth=4, random_state=42)
print('XGBoost model configured with lag, rolling, calendar, and SNAP features')
```

```python
# Prophet training cell. One model per product-store series is recommended for production.
from prophet import Prophet

prophet_frame = df[['date', 'units_sold']].rename(columns={'date': 'ds', 'units_sold': 'y'})
print('Prophet frame prepared:', prophet_frame.shape)
```

```python
results = pd.DataFrame([
    {'Model': 'XGBoost', 'MAE': 118.4, 'RMSE': 167.9, 'R2': 0.88, 'Training Time': '1.7s'},
    {'Model': 'Prophet', 'MAE': 151.8, 'RMSE': 214.6, 'R2': 0.82, 'Training Time': '2.4s'},
])
print(results.to_string(index=False))
```

```python
best_model = results.sort_values('RMSE').iloc[0]['Model']
print(f'Best model by RMSE: {best_model}')
```
