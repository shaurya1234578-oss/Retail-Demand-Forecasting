# Retail Demand Forecasting - EDA

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
DATA_PATH = Path('../data/sample_sales.csv')
df = pd.read_csv(DATA_PATH, parse_dates=['date'])
print(df.shape)
print(df.head(3).to_string(index=False))
```

```python
print(df.dtypes)
```

```python
print(df.isna().sum().to_string())
```

```python
daily = df.groupby('date', as_index=False)['units_sold'].sum()
print(daily.head().to_string(index=False))
daily.plot(x='date', y='units_sold', title='Daily Sales Trend')
```

```python
print(df['category'].value_counts().to_string())
```

```python
category_revenue = df.groupby('category')['revenue'].sum().sort_values(ascending=False)
print(category_revenue.to_string())
```

```python
city_units = df.groupby('city')['units_sold'].sum().sort_values(ascending=False)
print(city_units.to_string())
```

```python
promo_lift = df.groupby('promotion_flag')['units_sold'].mean().round(2)
print(promo_lift.to_string())
```
